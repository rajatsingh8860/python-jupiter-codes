```python
import matplotlib.pyplot as plt
import csv
x=[]
y=[]
with open ('test.txt','r') as csvfile:
    plot=csv.reader(csvfile)
    for col in plot:
        x.append(col[0])
        y.append(col[1])
plt.plot(x,y,label='file',color='blue')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('csv plot')
plt.legend()
plt.show()
```
